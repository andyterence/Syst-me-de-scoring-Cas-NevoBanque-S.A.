```python
#  Test du chi2 entre Payment_of_Min_Amount & Credit_Score

contingency_table = pd.crosstab(df["Payment_of_Min_Amount"], df["Credit_Score"])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"p-value : {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion : lien significatif entre Payment_of_Min_Amount et Credit_Score")
else:
    print("Conclusion : pas de lien significatif")
```
